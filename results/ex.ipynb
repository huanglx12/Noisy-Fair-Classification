```python
import pandas as pd
import numpy as np
```

```python
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import plotly.graph_objs as go
init_notebook_mode(connected=True)
```

```python
k1 = pd.read_csv('k_n11sigma10.txt', header=None,
                 names=['alg', 'error', 'prob', 'prob_std', 'gap'])
```

```python
k2 = pd.read_csv('k_n15sigma10.txt', header=None,
                 names=['alg', 'error', 'prob', 'prob_std', 'gap'])
```

```python
k3 = pd.read_csv('k_n19sigma10.txt', header=None,
                 names=['alg', 'error', 'prob', 'prob_std', 'gap'])
```

```python
k3
```

```python
#画prob图（画哪个文件的图，就用对应的kx，比如k1, k2....）
df = k3
x = df[df.alg == 0].error
y0 = df[df.alg == 0].prob
y1 = df[df.alg == 1].prob
y2 = df[df.alg == 2].prob

std0 = df[df.alg == 0].prob_std
std1 = df[df.alg == 1].prob_std
std2 = df[df.alg == 2].prob_std

gap0 = df[df.alg == 0].gap
gap1 = df[df.alg == 1].gap
gap2 = df[df.alg == 2].gap
```

```python
list(std0), list(std1), list(std2)
```

```python
gap1
```

```python
trace0 = go.Scatter(
    x = x,
    y = y0,
    name = 'Random (Prob)',
    mode = 'lines+markers',
    legendgroup = 'group0',
    marker = dict(color='rgb(65, 105, 225)'),
    error_y=dict(
        type='data',
        array=std0,
        thickness=1.5,
        width=3.5,
        visible=True
    )
)

trace1 = go.Scatter(
    x = x,
    y = y1,
    name = 'Rankorder (Prob)',
    mode = 'lines+markers',
    legendgroup = 'group1',
    marker = dict(color='rgb(255, 145, 0)'),
    error_y=dict(
        type='data',
        array=std1,
        thickness=1.5,
        width=3.5,
        visible=True
    )
)

trace2 = go.Scatter(
    x = x,
    y = y2,
    name = 'Our alg: Bucket (Prob)',
    mode = 'lines+markers',
    legendgroup = 'group2',
    marker = dict(color='rgb(46, 139, 87)'),
    error_y=dict(
        type='data',
        array=std2,
        thickness=1.5,
        width=3.5,
        visible=True
    )
)

trace3 = go.Scatter(
    x = x,
    y = gap0,
    name = 'Random (Gap)',
    legendgroup = 'group0',
    yaxis = 'y2',
    line=dict(dash='dash', width=2, color='rgb(65, 105, 225)')
    
)

trace4 = go.Scatter(
    x = x,
    y = gap1,
    name = 'Rankorder (Gap)',
    legendgroup = 'group1',
    yaxis = 'y2',
    line=dict(dash='dash', width=2, color='rgb(255, 145, 0)')
    
)

trace5 = go.Scatter(
    x = x,
    y = gap2,
    name = 'Our alg: Bucket (Gap)',
    legendgroup = 'group2',
    yaxis = 'y2',
    line=dict(dash='dash', width=2, color='rgb(46, 139, 87)')
    
)

layout1 = go.Layout(
    xaxis=dict(
        title='k',
        autorange=True,
        showgrid=True,
        zeroline=True,
        showline=True,
        ticks='',
        showticklabels=True
    ),
    yaxis=dict(
        title='success_prob',
        range=[0.5,1],
        showgrid=True,
        zeroline=True,
        showline=True,
        ticks='',
        showticklabels=True
    ),
    yaxis2=go.YAxis(
        title='Gap',
        range=[0.5,1],
        showline=True,
        side='right',
        overlaying="y",
    ),
    title='n=15, sigma=0.1',
    showlegend=True,
    legend=dict(x=1.12, y=1.05, tracegroupgap=5)
)

data = [trace0, trace1, trace2, trace3, trace4, trace5]
iplot(go.Figure(data=data, layout = layout1))
```
